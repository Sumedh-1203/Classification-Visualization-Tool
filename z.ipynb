```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('Dataset1/train.csv')
df1.head()
```

```python
plt.scatter(df1['x1'], df1['x2'], c=df1['label'])
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar(label='label')
plt.show()
```
